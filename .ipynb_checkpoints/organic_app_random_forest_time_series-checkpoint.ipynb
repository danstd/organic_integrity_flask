```python
# Transform the trend data into a format for supervised learning.
trend = trend_df.Trend

# Remove the current month from the data, as the full count is not in place.
c_y = str(datetime.datetime.now().year)
c_m = str(datetime.datetime.now().month)
c_start_m = c_y + "-" + c_m + "-01"
current = datetime.datetime.strptime(c_start_m, "%Y-%m-%d")

trend = trend.loc[trend.index < current]

# Get the trends column and the shifted data column into a new dataframe. Extract values.
trends = pd.DataFrame(trend)
```

```python
# Group by month.
trends = trends.reset_index()

trends["Month"] = trends.op_statusEffectiveDate.apply(lambda x: datetime.datetime.strftime(x, "%Y-%m"))

trends = trends.groupby("Month", as_index=False)["Trend"].sum()

trends["Year"] = trends["Month"].apply(lambda x: x.split("-")[0])

trends["Month"] = trends["Month"].apply(lambda x: x.split("-")[1])
```

```python
trends["Shift_Back_1"] = trends["Trend"]

trends.Shift_Back_1 = trends.Shift_Back_1.shift(1)

trends["Shift_Back_2"] = trends.Shift_Back_1.shift(1)

trends["Shift_Back_3"] = trends.Shift_Back_1.shift(2)
```

```python
# Reorganize the columns.
trends = trends[["Year", "Month", "Shift_Back_3", "Shift_Back_2", "Shift_Back_1", "Trend"]]
```

```python
trends = trends.dropna()
trends
```

```python
trend_vals = trends.values
```

```python
# Split the dataset into training, validation, and testing sets.
# Using ~20% for validation and ~10% for testing.
val_length = math.ceil(len(trend_vals) * 0.2)
test_length = math.ceil(len(trend_vals) * 0.1)
train_length = len(trend_vals) - (val_length + test_length)

train = trend_vals[0:train_length]

validation = trend_vals[train_length: (train_length + val_length)]
validation_set = trends[train_length: (train_length + val_length)]

test = trend_vals[(train_length + val_length):]
test_set = trends[(train_length + val_length):]
```

```python
train_orig = train
validation_orig = validation
```

```python
# Create random forest models using provided criteria.
# Modified from: https://machinelearningmastery.com/random-forest-for-time-series-forecasting/
def r_forest_iter(params, start_cols):
    score_list = list() # Hold score against validation.
    model_list = list() # Hold models
    pred_list = list() # hold list of predictions.
    param_output = list() # hold the parameters used, for reference.
    start_col_output = list() # Hold the start column for features, for reference.
    for param in range(0,len(params)):
        for start_col in start_cols:
            train = train_orig
            validation = validation_orig
            history = [i for i in train]
            predictions = list()
            for i in range(len(validation)):
                # split test row into input and output columns
                val_x, val_y = validation[i, start_col:-1], validation[i, -1]
                # fit model on history and make a prediction

                # transform list into array
                train = np.asarray(history)
                # split into input and output columns
                train_x, train_y = train[:, start_col:-1], train[:, -1]

                # fit model
                model = ske.RandomForestRegressor(n_estimators=params[param]["n-estimators"], criterion=params[param]["criterion"])
                model.fit(train_x, train_y)

                # make a one-step prediction
                pred_result = model.predict([val_x])[0]

                # store forecast in list of predictions
                predictions.append(pred_result)

                # add actual observation to history for the next loop
                history.append(validation[i])

            # Get MAE and RMSE
            score = list()
            score.append(skm.mean_absolute_error(validation[:,-1], predictions))
            score.append(math.sqrt(skm.mean_squared_error(validation[:,-1], predictions)))
            
            param_output.append(param)
            start_col_output.append(start_col)
            score_list.append(score)
            pred_list.append(predictions)    
            model_list.append(model)
    return param_output, start_col_output, score_list, model_list, pred_list
```

```python
# Make a list of dictionaries with different parameters for the model.
model_params = list()
model_params.append({"n-estimators":500, "criterion": "absolute_error"})
model_params.append({"n-estimators":1000, "criterion": "absolute_error"})
model_params.append({"n-estimators":500, "criterion": "squared_error"})
model_params.append({"n-estimators":1000, "criterion": "squared_error"})
```

```python
start_cols = [0, 1, 2, 3, 4]
```

```python
params, start_cols, scores, models, predictions = r_forest_iter(model_params, start_cols)
```

```python
scores = pd.DataFrame(scores, columns=["MAE", "RMSE"])
params_and_cols = pd.DataFrame(list(zip(params,start_cols)), columns=["Params", "Start_cols"])
scores.merge(right=params_and_cols, left_index=True, right_index=True)
```

```python
results = scores.merge(right=params_and_cols, left_index=True, right_index=True)
```

```python
# Get the first of each month from a given starting timestamp for the number in months.
def month_accumulate(start_month, num_months):
    month_list = list()
    temp = start_month
    for i in range(0,num_months):
        next_month = temp + datetime.timedelta(days=32)
        day_subtract = next_month.day - 1
        next_month = next_month - datetime.timedelta(days=day_subtract)
        month_list.append(next_month)
        temp = next_month
    return month_list
```
